```python
import requests
import json

city = "Seoul"
apikey = "b486db75596d5d02db8994fa4e6203bf"
lang = "kr"

api = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={apikey}&lang={lang}&units=metric"

result = requests.get(api)
data = json.loads(result.text)

print(data["name"],"의 날씨입니다.")
print("날씨는 ",data["weather"][0]["description"],"입니다.")
print("현재 온도는 ",data["main"]["temp"],"입니다.")
print("하지만 체감 온도는 ",data["main"]["feels_like"],"입니다.")
print("최저 기온은 ",data["main"]["temp_min"],"입니다.")
print("최고 기온은 ",data["main"]["temp_max"],"입니다.")
print("습도는 ",data["main"]["humidity"],"입니다.")
print("기압은 ",data["main"]["pressure"],"입니다.")
print("풍향은 ",data["wind"]["deg"],"입니다.")
print("풍속은 ",data["wind"]["speed"],"입니다.")
```

## 링크 만들기

## Openweathermap

API doc == 사용 방법 설명서

지역 이름으로 날씨를 받아오는 API

"http://api.openweathermap.org/data/2.5/weather?q={city name}&appid={API key}"

중괄호 {}로 묶여 있는 부분은 사용자가 직접 지정해줘야한다. 이런것들을 파라미터라고 부른다.

```python
city = "Gwangju"
apikey = "b486db75596d5d02db8994fa4e6203bf"
lang = "kr"   # 언어 설정 # &lang={lang} 언어 설정
units= "metric" # 화씨 -> 섭씨 변경
api = f"""http://api.openweathermap.org/data/2.5/weather?q={city}&appid={apikey}&lang={lang}&units={units}"""
```

```python
# api 서버에서 response 해준다.
# 서버 주소의 의미
# ?를 기준으로 앞쪽은 공통 url 뒤쪽은 파라미터이다.
print(api)
```

# 날씨 받아오기

### json == javascript object notation

### json == 데이터를 주고 받을 때 사용하는 포맷

### json == 파이썬이 기본적으로 제공하는 모듈

json은 dictionary 형태를 가지고 있는 문자열이다.

```python
import requests # 요청 보내기
import json

result = requests.get(api)
print(result.text)

data = json.loads(result.text) #== json 타입으로 변환

for i in data:
    print(i)
```
